```python
import glob
```

```python
def get_freqs_masked(grp='A'):
    fnames = glob.glob('/data2/GBT/*/*/*/*.'+grp+'.*rfireduced.npz')
    fnames.sort()
    N = 2**14
    tot_index = np.zeros(N)
    for f in fnames:
        x = np.load(f)
        z = np.zeros(N)
        z[~x['mask']] = 1
        tot_index += z
    df = x['CDELT1'][0]
    center_freq = x['CRVAL1'][0]
    centerbin = x['CRPIX1'][0]
    index = np.arange(N)
    freqs = (center_freq + df * (index + 1 - centerbin))[::-1]
    return freqs, tot_index
```

```python
freqsA, indexsA = get_freqs_masked('A')
```

```python
freqsB, indexsB = get_freqs_masked('B')
```

```python
freqsC, indexsC = get_freqs_masked('C')
```

```python
freqsD, indexsD = get_freqs_masked('D')
```
