```python
figsize(16, 10)
ghz = 1.e-9
```

```python
x = np.load('highnoise_diode_calibration_3C286_01.npy')
x1 = np.load('highnoise_diode_calibration_3C295_01_0.npy')
x2 = np.load('highnoise_diode_calibration_3C295_01_1.npy')
plot(x[0]*ghz, x[1], label='286')
plot(x1[0]*ghz, x1[1], label='295 0')
plot(x2[0]*ghz, x2[1], label='295 1')
xlabel('Frequency [GHz]')
ylabel('High noise Diode [Jy]')
legend()
```
