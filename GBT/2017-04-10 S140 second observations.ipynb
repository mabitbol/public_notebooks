```python
from astropy.io import fits
import glob
```

```python
antenna_files = glob.glob('/home/gbtdata/AGBT17A_259_03/Antenna/2017_04_10_*.fits')
antenna_files.sort()
antenna_files = antenna_files[7:]
#antenna_files += glob.glob('/home/gbtdata/AGBT17A_259_02/Antenna/2017_04_05_0[89]*.fits')
#antenna_files.sort()
antenna_files2 = glob.glob('/home/gbtdata/AGBT17A_259_05/Antenna/2017_*.fits')
antenna_files2.sort()
antenna_files += antenna_files2[7:]
ra = []
dec = []
for fn in antenna_files:
    try:
        f = fits.open(fn)
    except:
        continue
    h = f[2]
    ra.extend(h.data['RAJ2000'])
    dec.extend(h.data['DECJ2000'])
    f.close()
ra = np.array(ra)
dec = np.array(dec)
```

```python
bins_ra = np.arange(ra.min(), ra.max(), 1.8 / 60.)
bins_dec = np.arange(dec.min(), dec.max(), 1.8 / 60.)
stuff = np.histogram2d(ra, dec, bins=[bins_ra, bins_dec])
histo = stuff[0]
```

```python
figure(figsize=(8, 8))
pck = pcolormesh(bins_ra, bins_dec, histo.T, cmap=cm.viridis)
pck.cmap.set_under('grey')
clim(1., 10)
colorbar()
```
