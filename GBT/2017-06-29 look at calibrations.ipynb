```python
ghz = 1.e-9
```

```python
df = 0.75
df2 = 0.3
freqs = [4.575, 5.625, 6.125, 7.175]
```

```python
figsize(10, 8)
#plot(fs, s1p0, 'r', label='S1')
#plot(fs, s1p1, 'r')
plot(fs*ghz, s2p0, 'r', label='Session 2')
plot(fs*ghz, s3p0, 'b', label='Session 3')
plot(fs*ghz, s4p0, 'g', label='Session 4')
plot(fs*ghz, s4p1, 'g')
plot(fs*ghz, s4p2, 'g')
plot(fs*ghz, s5p0, 'cyan', label='Session 5')
plot(fs*ghz, s5p1, 'cyan')

ylabel('Diode power [Jy]', fontsize=20)
legend(loc=2, ncol=2, fontsize=20, framealpha=0)
xlabel('Frequency [GHz]', fontsize=20)
title('Diode power calibrated on 3C295', fontsize=20)
xticks(fontsize=20)
yticks(fontsize=20)
grid()

axvspan(freqs[0]-df, freqs[0]+df, alpha=0.1, color='grey')
axvspan(freqs[1]-df, freqs[1]+df, alpha=0.1, color='grey')
axvspan(freqs[2]-df, freqs[2]+df, alpha=0.1, color='grey')
axvspan(freqs[3]-df, freqs[3]+df, alpha=0.1, color='grey')

text(freqs[0]-df2, 2.05, 'Bank A', fontsize=16)
text(freqs[1]-df2, 2.05, 'Bank B', fontsize=16)
text(freqs[2]-df2, 2.05, 'Bank C', fontsize=16)
text(freqs[3]-df2, 2.05, 'Bank D', fontsize=16)

savefig('calibrations_3C295_all.pdf', fmt='pdf')
```
